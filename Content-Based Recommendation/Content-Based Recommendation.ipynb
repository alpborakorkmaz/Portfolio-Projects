```python
import pandas as pd
pd.set_option('display.max_columns', None)
pd.set_option('display.width', 500)
pd.set_option('display.expand_frame_repr', False)
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
df = pd.read_csv("movies_metadata.csv", low_memory=False)  
```

```python
df.head()
```

```python
# to delete the most commonly used words such as the, and, on,..
tfidf = TfidfVectorizer(stop_words="english")
# to fill missing values with " "
df['overview'] = df['overview'].fillna('')
# applying tfidf method to overview variable to get the matrix
tfidf_matrix = tfidf.fit_transform(df['overview'])
```

# Cosine-Sim Calculator

```python
cosine_sim = cosine_similarity(tfidf_matrix,
                               tfidf_matrix)
```

# Recommendation Based on Similarities

```python
indices = pd.Series(df.index, index=df['title'])
indices = indices[~indices.index.duplicated(keep='last')]
```

```python
movie_index = indices["Sherlock Holmes"]
cosine_sim[movie_index]
similarity_scores = pd.DataFrame(cosine_sim[movie_index],
                                 columns=["score"])
movie_indices = similarity_scores.sort_values("score", ascending=False)[1:11].index
df['title'].iloc[movie_indices]
```
